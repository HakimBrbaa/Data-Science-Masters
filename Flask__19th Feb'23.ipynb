#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight web application framework written in Python that is designed to make building web applications easier and more efficient. Flask allows developers to create web applications quickly and easily, without the need for boilerplate code or complicated configurations.

One advantage of Flask is its flexibility. Flask allows developers to create web applications in a way that suits their needs and preferences. Developers can choose which components they want to use, and they can configure these components in a way that works best for their application.

Another advantage of Flask is its simplicity. Flask has a small and easy-to-understand codebase, which makes it easy for developers to learn and use. Flask's simplicity also makes it easy to maintain and extend, which can save developers time and effort in the long run.

Flask is also highly customizable, which means that developers can easily add new features and functionality to their web applications. Flask's modular design allows developers to add new components or modify existing ones, without affecting the rest of their application.

Finally, Flask has a large and active community of developers, which means that there are many resources and tools available for developers who use Flask. This community provides support, documentation, and examples, which can help developers learn and use Flask more effectively.

#### Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping a URL endpoint to a specific function in a Flask application. When a client makes a request to a particular URL endpoint, Flask uses app routing to determine which function should handle the request and generate the appropriate response.

App routes are used in Flask for several reasons, including:

- Routing requests to the appropriate view function: App routing allows developers to map specific URLs to the view functions that should handle them. This makes it easy to create a modular and organized application structure, where each view function handles a specific set of requests.

- Creating RESTful APIs: App routing is essential for building RESTful APIs in Flask. By mapping each URL endpoint to a specific function that handles the corresponding HTTP method (e.g. GET, POST, PUT, DELETE), Flask applications can easily expose an API that follows the RESTful design principles.

- Enabling URL parameters: App routing allows developers to define dynamic URL endpoints that accept parameters. For example, a URL endpoint might include a parameter for a specific user ID or product SKU. By using app routing, Flask can extract the parameter from the URL and pass it as an argument to the corresponding view function.

Overall, app routing is a fundamental part of building web applications in Flask. By using app routes, developers can create a flexible and modular application structure that can handle a wide range of requests and support various types of functionality.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The 'url_for()' function in Flask is used for URL building. It generates a URL for a given endpoint, taking into account any arguments that need to be passed in the URL. This function is especially useful because it allows developers to create dynamic URLs that can be easily changed without affecting the rest of the application.

Here is an example Python code that demonstrates the use of the url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/welcome")
def welcome_message():
    return "Welcome to ABC Corporation"

@app.route("/")
def welcome_message1():
    a = {
        "Company Name" : "ABC Corporation",
        "Location" : "India",
        "Contact Detail" : "999-999-9999"
        }
    return a
with app.test_request_context():
    # Use url_for to generate a URL for the welcome_message function
    print(url_for('welcome_message'))

    # Use url_for to generate a URL for the welcome_message1 function
    print(url_for('welcome_message1'))
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
